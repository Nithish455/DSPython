#Identifying outliners using IQR:

```python
a.isna().sum()
```

```python
import numpy as np
a['Volume'].isna().sum()
vol = a['Volume'].dropna()
q1 = vol.quantile(0.25)
q3 = vol.quantile(0.75)
iqr = q3-q1
lower_bound = q1 - (1.5*iqr)
upper_bound = q3 + (1.5*iqr)
outliers = vol[(vol<lower_bound) | (vol>upper_bound)]
print("Outliers:",outliers)
```

#Create a new column in past dataset:

```python
a['dates'] = pd.to_datetime(a['Date']).dt.year.apply(lambda x: 'old' if x <= 1988 else 'new')
print(a[['Date','dates']].tail())
```
