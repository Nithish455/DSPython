# Step-2: Import packages

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

# Step-3:Load Dataset

```python
a = pd.read_csv("/content/Project-4_training_data_with_weather_info_week_1.csv")
```

```python
#shape
a.shape
```

```python
#Null Values
a.isnull().sum()
```

```python
#fill null values
#Method-1
a.fillna(0,inplace=True)
```

```python
#Method-2
#we prefer median to fill int values but we cannot fill float values using median so we use mode to fill the data
a['Province/State'].fillna(a['Province/State'].mode()[0],inplace = True)
a['min'].fillna(a['min'].mode()[0],inplace = True)
a['max'].fillna(a['max'].mode()[0],inplace = True)
a['slp'].fillna(a['slp'].mode()[0],inplace = True)
a['dewp'].fillna(a['dewp'].mode()[0],inplace = True)
a['rh'].fillna(a['rh'].mode()[0],inplace = True)
a['ah'].fillna(a['ah'].mode()[0],inplace = True)
```

```python
a.isna().sum()
```
